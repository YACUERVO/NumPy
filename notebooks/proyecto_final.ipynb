Objetivo 

El objetivo es analizar y modelar la tendencia de crecimiento en el numero de pasajeros de aerolineas internacionales entre 1949 y 1960, utilizando un modelo matematico.

```python
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt(url, delimiter=',', skip_header=1, usecols=1)
```

```python
plt.figure(figsize=(10,6))
plt.plot(data, label='Datos de pasajeros',linewidth=2,linestyle='--',marker='o')
plt.title('Datos de pasajeros de la aerolinea (1949-1960)')
plt.xlabel('Año')
plt.ylabel('Pasajeros')
plt.legend()
plt.grid(True)
plt.show()
```

```python
log_data = np.log(data)
x = np.arange(1,len(data)+1)
X = np.column_stack((np.ones(len(x)),x))
x
```

```python
X
```

```python
#Modelado exponencial
# Este código realiza el modelado exponencial de los datos de pasajeros de aerolíneas.
# Primero, se calcula la pseudoinversa de la matriz X, que contiene una columna de unos (para el término independiente) y una columna con los valores de x (los meses).
# Luego, se multiplica la pseudoinversa de X por los datos transformados con logaritmo natural (log_data), para obtener los parámetros del modelo lineal en el espacio logarítmico.
# Estos parámetros corresponden a log(A) (el intercepto) y B (la pendiente) en el modelo logarítmico: log(y) = log(A) + Bx,
# que equivale a y = A * exp(Bx) en el espacio original.
# Finalmente, se imprime el vector de parámetros obtenidos.


x_pinv = np.linalg.pinv(X)
params = np.dot(x_pinv,log_data)
print(params)
```

```python
# Este código realiza el modelado exponencial de los datos de pasajeros de una aerolínea.
# Primero, se calcula la pseudoinversa de la matriz X, que contiene una columna de unos (para el término independiente) y una columna con los valores de x (los meses).
# Luego, se multiplica la pseudoinversa de X por los datos transformados con logaritmo natural (log_data), para obtener los parámetros del modelo lineal en el espacio logarítmico.
# Estos parámetros corresponden a log(A) (el intercepto) y B (la pendiente) en el modelo logarítmico: log(y) = log(A) + Bx,
# que equivale a y = A * exp(Bx) en el espacio original.
# Finalmente, se imprime el vector de parámetros obtenidos.

log_A = params[0]
B = params[1]
A = np.exp(log_A)

```

```python
#Presentacion de resultados
fitted_values = A * np.exp(B * x)
plt.figure(figsize=(10,6))
plt.plot(data, label='Datos de pasajeros',linewidth=1,linestyle='--',marker='o')
plt.plot(fitted_values, label='Ajuste exponencial',linestyle='--',marker='o',color='red')
plt.title('Datos de pasajeros de la aerolinea (1949-1960)')
plt.xlabel('Año')
plt.ylabel('Pasajeros')
plt.legend()
plt.grid(True)
plt.show()
```
